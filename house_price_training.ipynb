```python
#importing the necessary libraries 

import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
import pickle

# loading the dataset("House_prices") 
data_set=pd.read_csv("D:\\machine learning\\python_practice\\House_Prices.csv")

#converting the dataset into dataframe

df=pd.DataFrame(data_set)

#converting the categorial into the numerical 

# brick column(using the map)

df["Brick"]=df["Brick"].map({'Yes':1,'No':0})

# Neighbourhood column (one hot encoding )
df=pd.get_dummies(df,columns=["Neighborhood"])


#splitting the input fetaure and output features
y=df["Price"]
x=df.drop(["Home","Price"],axis=1)


# splitting the dataset into training set and testing set

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#training the model with training data

model=LinearRegression()
model.fit(x_train,y_train)

#evalute the model performace
prediction=model.predict(x_test)

#calculating the mean squarred error and r2 error
mse=mean_squared_error(y_test,prediction)
r=r2_score(y_test,prediction)
print(mse)
print(r)

# dumping the model


pickle.dump(model, open("house_price_model.pkl", "wb"))
```
